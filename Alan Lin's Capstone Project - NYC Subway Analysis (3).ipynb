# Part 2. Turnstile Analysis

## 1. Load the data

Description:
* C/A = Control Area (e.g., A002)
* unit = Remote Unit for a station (e.g., R051)
* SCP = Subunit Channel Position represents an specific address for a device (e.g., 02-00-00)
* station_code = C/A + unit, locating a station
* turnstile = C/A + unit + SCP, locating a turnstile
* Station = Represents the station name the device is located at
* date = Represents the date (MM-DD-YY)
* time = Represents the time (hh:mm:ss) for a scheduled audit event
* datetime = date + time (MM-DD-YY hh:mm:ss)
* desc = Represent the "REGULAR" scheduled audit event (Normally occurs every 4 hours)
* entries = The comulative entry register value for a device
* exits = The cumulative exit register value for a device

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import seaborn as sns
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
```

```python
turnstile = pd.read_csv('turnstile_170401.csv')
```

```python
turnstile
```

```python
# before convert datetime format, checking the unique value for date
turnstile['DATE'].unique()
```

```python
turnstile['DATETIME'] = pd.to_datetime(turnstile.DATE + ' ' + turnstile.TIME, format='%m/%d/%Y  %H:%M:%S')
turnstile['turnstile'] = turnstile['C/A'] + '-' + turnstile['UNIT'] + '-' + turnstile['SCP']

turnstile = turnstile[(turnstile.DATETIME >= '01-01-17 00:00:00') & 
          (turnstile.DATETIME <'07-01-17 00:00:00')]
```

* Perform Hypothesis Test to test which features do depend on Entries or Exits

```python
import scipy.stats as stats
# T-test to check whether Hour depend on Entries or not
Ho = "Hour does not depend on Entries"   # Stating the Null Hypothesis
H1 = "Hour depends on Entries"   # Stating the Alternate Hypothesis

x = np.array(turnstile_weather['Hour'])  
y = np.array(turnstile_weather['ENTRIES']) 

t, p_value  = stats.ttest_ind(x,y, axis = 0)  #Performing an Independent t-test
```

```python
print("The p value is", p_value)
```

```python
print("The t value is", t)
```

```python
if p_value < 0.05:  # Setting our significance level at 5%
    print(f'{H1} as the p_value ({p_value}) < 0.05 and Reject the Null Hypothesis')
else:
    print(f'{Ho} as the p_value ({p_value}) > 0.05 and Accept the Null Hypotheis')
```
